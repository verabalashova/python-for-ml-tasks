**1. Функція з параметром**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
def show_menu(food_list):
    """
    Метод, що за отрманим списком страв виводить форматоване меню
    
    :param food_list: список страв
    """
    print("=== МЕНЮ ===")
    for i, food_item in enumerate(food_list):
        print(f"{i + 1}. {food_item}")
    print("=============")

monday_dishes = ['Піца', 'Паста', 'Салат'];
tuesday_dishes = ["Суп грибний", "Котлета по-київськи", "Лимонад"]

show_menu(monday_dishes)
print()
show_menu(tuesday_dishes)
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
    """
    Метод, що виводить квадрат числа у повідомлення форматованого вигляду
    
    :param n: число, квадрат якого буде вивеено 
    """
    print(f"Квадрат числа {n} = {n ** 2}")

print_square(5)
print_square(-5)
print_square(0)
print_square(16)
print_square(-45)
print_square(12.5)
````

**3. Функція з поверненням значення**

Створіть функцію `multiply(a, b)`, яка повертає добуток двох чисел.

Викличте функцію на власноруч заданих значеннях та виведіть результат.


````python
def multiply(a, b):
    """
    Функція, що повертає добуток двох чисел
    
    :param a: множник 1
    :param b: множник 2 
    """
    return a * b

x = 2
y = 4
print(f"Добуток {x} і {y} = {multiply(x, y)}" )

x = -11
y = 5
print(f"Добуток {x} і {y} = {multiply(x, y)}")

x = 13.75
y = 13
print(f"Добуток {x} і {y} = {multiply(x, y)}")
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
def find_max(numbers):
    """
    Функція, що повертає найбільше число в списку
    
    :param numbers: список чисел
    """
    return max(numbers)

numbers_list = [15, 42, 8, 23, 56, 19]
print(f" Найьільше число для списку {numbers_list} це {find_max(numbers_list)}")
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
import random

def check_grade(grade):
    """
    Функція, що повертає рядок "Здав", якщо grade >= 60 або "Не здав" в протилежному випадку
    
    :param grade: оцінка студента
    """
    return "Здав" if grade >= 60 else "Не здав"

random.seed(29)
random_grages = [random.randint(30, 100) for _ in range(15)]
for i, random_grade in enumerate(random_grages):
    print(f"{i + 1}. Студент з оцінкою {random_grade}: ", check_grade(random_grade))
````
