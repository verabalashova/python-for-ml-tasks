**3. Фільтрація даних**

Відфільтруйте повний набір даних та виведіть (кожен пункт оцінюється в 0.5 бала):
- Всі рахунки більше 20 доларів (`total_bill > 20`)
- Всі відвідувачі-жінки (`sex == 'Female'`)
- Рахунки на вечерю (`time == 'Dinner'`)
- Рахунки на вечерю з сумою чеку більше 30 доларів

Для виведення результатів фільтрації:
- зберегти результат у окрему змінну
- показати кількість записів у відфільтрованому наборі
- вивести перші 5 записів відфільтрованого набору.

```python
def display_filter_result(title_message: str, df_param: pd.core.frame.DataFrame):
    print(title_message.upper())
    print(f'Кількість записів у відфільтрованому наборі: {df_param.size}')
    display(df_param.head())
    print()


# - Всі рахунки більше 20 доларів (`total_bill > 20`)
all_bills_higher_than_20 = df[(df['total_bill'] > 20)]
display_filter_result('1. Всі рахунки більше 20 доларів', all_bills_higher_than_20)

# - Всі відвідувачі-жінки (`sex == 'Female'`)
all_female_visitors = df[(df['sex'] == 'Female')]
display_filter_result('2. Всі відвідувачі-жінки', all_female_visitors)

# - Рахунки на вечерю (`time == 'Dinner'`)
all_dinner_bills = df[(df['time'] == 'Dinner')]
display_filter_result('3. Рахунки на вечерю', all_dinner_bills)

# - Рахунки на вечерю з сумою чеку більше 30 доларів
all_dinner_bills_higher_than_30 = df[(df.time == 'Dinner') & (df.total_bill > 30 )]
display_filter_result('4. Рахунки на вечерю з сумою чеку більше 30 доларів', all_dinner_bills_higher_than_30)
```

**4. Описова статистика**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній розмір рахунку на всьому наборі даних
- Максимальні чайові
- Кількість пропущених значень у всіх колонках
- Скільки в наборі даних рахунків за обіди і за вечері

```python
# - Середній розмір рахунку на всьому наборі даних
mean_tip = df['tip'].mean()
print(f'1. Середній розмір рахунку на всьому наборі даних: {mean_tip:.2f} (non-formatted: {mean_tip})')
print()

# - Максимальні чайові
max_tip = df.tip.max()
print(f'2. Максимальні чайові: {max_tip:.2f} (non-formatted: {max_tip})')
print()

# - Кількість пропущених значень у всіх колонках
print(f'3. Кількість пропущених значень у всіх колонках:')
display(df.isnull().sum())
print()

# - Скільки в наборі даних рахунків за обіди і за вечері
df['is_dinner'] = (df['time'] == 'Dinner').astype(int)
df['is_lunch'] = (df['time'] == 'Lunch').astype(int)

print('4. В наборі даних рахунків:')
print(f' - за обіди: {df.is_lunch.sum()}')
print(f' - за вечері: {df.is_dinner.sum()}')
print(f' - за обіди і за вечері: {df.is_lunch.sum() + df.is_dinner.sum()}')
```

**5. Статистика із групуванням**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній рахунок за статтю (groupby `sex`)
- Середні чайові за днем тижня (groupby `day`)
- Скільки відвідувачів у кожен день

```python
# - Середній рахунок за статтю (groupby `sex`)
mean_by_sex = df.groupby('sex')['total_bill'].mean()
print('1. Середній рахунок за статтю:')
display(mean_by_sex.round(2))
print()

# - Середні чайові за днем тижня (groupby `day`)
mean_tips_by_day = df.groupby('day')['tip'].mean()
print('2. Середні чайові за днем тижня:')
display(mean_tips_by_day.round(2))
print()

# - Скільки відвідувачів у кожен день
visitor_stats_by_day = df.groupby('day').agg({
    'size': ['count', 'sum']
})
print('3. Скільки відвідувачів у кожен день (платники і загальна кількість):')
display(visitor_stats_by_day)
print()
```

**6. Створення нової колонки**

Створіть нову колонку `tip_percentage` - відсоток чайових від загальної суми рахунку.
Виведіть перші 10 рядків з новою колонкою.

```python
df['tip_percentage'] = df['tip'] / df['total_bill'] * 100
print('Відсоток чайових від загальної суми рахунку:')
display(df['tip_percentage'].round(2).head(10))
```

**7. Сортування**

Відсортуйте датасет і виведіть результат сортування в кожному випадку окремо (кожен пункт оцінюється в 0.5 бала):
- За розміром рахунку (зростання)
- За чайовими (спадання)

Додатково виведіть топ-5 найбільших чайових використовуючи метод `nlargest`.

```python
# - За розміром рахунку (зростання)
df_sorted_by_total_bill_asc = df.sort_values('total_bill')
print('1. За розміром рахунку (зростання)'.upper())
display(df_sorted_by_total_bill_asc)
print()

# - За чайовими (спадання)
df_sorted_by_tips_desc = df.sort_values('tip', ascending=False)
print('2. За чайовими (спадання)'.upper())
display(df_sorted_by_tips_desc)
print()

# - топ-5 найбільших чайових використовуючи метод `nlargest`.
top_tips = df.nlargest(5, 'tip')
print('3. топ-5 найбільших чайових використовуючи метод `nlargest`'.upper())
display(top_tips)
print()
```
