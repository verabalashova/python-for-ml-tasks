**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
import numpy as np
```

```python
# Масив зі списку `[5, 15, 25, 35, 45]`
array_from_list = np.array([5, 15, 25, 35, 45])
print("1. Масив зі списку:", array_from_list, ". Тип результату: ", type(array_from_list))

# Масив з нулів розміром 7
zero_array = np.zeros(7, dtype=int);
print("2. Масив з нулів розміром 7:", zero_array, ". Тип результату: ", type(zero_array))

# Масив з одиниць розміром 6
ones_array = np.ones(6, dtype=int);
print("3. Масив з одиниць розміром 6:", ones_array, ". Тип результату: ", type(ones_array))

# Масив з числами від 10 до 19 використовуючи `np.arange()`
custom_array = np.arange(10, 20, 1)
print("4. Масив з числами від 10 до 19 використовуючи `np.arange()`:", custom_array, ". Тип результату: ", type(custom_array))
```
